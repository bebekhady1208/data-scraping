# **BEAUTIFULSOUP VENTE MOTO  SCRAPING**

```python
# Importation des packages requis
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
from requests import get
```

```python
# Initialisation d'un DataFrame vide
df = pd.DataFrame()
```

```python
# Parcoure des 101 premières pages

# for p in range(1, 101) #la ligne normale
for p in range(1, 3):
    url = f'https://dakar-auto.com/senegal/motos-and-scooters-3?page={p}'
    res = requests.get(url)
    soup = bs(res.text, 'html.parser')
    containers = soup.find_all('div', class_='listing-card')

    data = []

    for container in containers:
        name = container.find('h2', class_='listing-card__header__title mb-md-2 mb-0').text.strip()
        price = container.find('h3', class_='listing-card__header__price font-weight-bold text-uppercase mb-0').text.strip().replace('\u202f','').replace(' F CFA','')
        Ref = container.find('li', class_='listing-card__attribute list-inline-item').text.strip().replace('\u202f','').replace('Ref.','')
        proprio = container.find('p', class_='time-author m-0').text.strip().replace('\u202f','').replace('Par ','')
        adr_vente_commune = container.find('span', class_='town-suburb d-inline-block').text.strip().replace('\u202f','').replace(',','')
        adr_vente_region = container.find('span', class_='province font-weight-bold d-inline-block').text.strip()
        kilometrage = container.find('i',class_='v-icon icon-road-perspective')





        data.append({'Marque_Modele_Annee': name,
                     'Prix (F CFA)' : price,
                     'Ref.' : Ref,
                     'Proprio' : proprio,
                     'Adr_Commune' : adr_vente_commune,
                     'Adr_Region' : adr_vente_region,
                     'Kilometrage' : kilometrage,
                     })

    # Append data to the DataFrame
    df = pd.DataFrame(data)

#Print the DataFrame
print(df)
```

```python
# Reorganisation des colonnes

df = df[['Marque_Modele_Annee', 'Kilometrage', 'Prix (F CFA)', 'Ref.', 'Adr_Commune', 'Adr_Region', 'Proprio' ]]
```

```python
# Divisions de la colonne 'Marque_Modele_Annee' partir des espaces
split_columns = df['Marque_Modele_Annee'].str.split(' ', expand=True)
```

```python
# Attribution des colonnes fractionnées aux nouvelles colonnes de noms : 'Marque' , 'Model', 'Annee'
df['Marque'] = split_columns[0]
df['Model'] = split_columns[1]
df['Annee'] = split_columns[2]
```

```python
# Nouvellle reorganisation des colonnes
df = df[['Marque', 'Model', 'Annee', 'Kilometrage', 'Prix (F CFA)', 'Ref.', 'Adr_Commune', 'Adr_Region', 'Proprio' ]]
```
